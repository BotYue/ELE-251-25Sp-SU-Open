<div style="border-left: 4px solid #28a745; padding: 10px; background-color: #f9fff3; margin-top: 10px;"><strong>Task 1.1 NumPy Array</strong> 
</div>

Consider that you have a 7-day temperature of Syracuse in Fahrenheit degree $\degree F$:

26, 25, 34, 37, 23, 14, 9, 14 

Now you wan to convert them to Celsius Degree $\degree C$.

In Python, without any library, you may need to write a Python loop to compute it. 

For example:

```python
# This is a Python Code cell. Run it
temps_fahrenheit = [26, 25, 34, 37, 23, 14, 9, 14]

temps_celsius = [(temp - 32) / 1.8 for temp in temps_fahrenheit]

print(temps_celsius)
```

<span style="border-left: 4px solid #FFA500; padding: 5px; background-color: #FFF4E5; display: inline-block;">
    <strong>Python output: 0.5 pts</strong>
</span>

If we use NumPy library and store the temperatures in a NumPy Array, it will be easier to code:

```python
# This is a Python Code cell. Run it
import numpy as np

temps_fahrenheit = np.array([26, 25, 34, 37, 23, 14, 9, 14])

temps_celsius = (temps_fahrenheit - 32) / 1.8

print(temps_celsius)
```

    <strong>Python output: 0.5 pts</strong>
</span>

<div style="border-left: 4px solid #28a745; padding: 10px; background-color: #f9fff3; margin-top: 10px;"><strong>Task 2.1 Plot a continuous line</strong> 
</div>

Let's consider a simple case:

Plotting quadratic equation $ y = -x^2 + 9x +100$ in a $xy$ plane.

To do so, we typically do the following steps:

* Create a NumPy array of $x$ values.
  Use ``np.linspace(  ,  , )``. <br> The 3 entries are start point, end point, numbers of points.
* Evaluate the equation using created array.
* Use ``plt.plot(x, y)`` to generate plot.

Lets have a try.

```python
# This is a Python Code cell. Run it
x = np.linspace(-10, 10, 5)
y = -x**2 + 9*x + 100
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
```

After running the code, you may notice it is not Smooth.  
This issue is due to the 3rd entry of the np.linspace(-10, 10, 5).

NumPy only evaluates at 5 points.  
So Matplotlib plots these 5 points and then automatically connects them with straight lines, resulting in a less smooth curve.

Now, in Python Code cell below, adjust the 3rd entry of ``np.linspace(  ,  , )`` to a larger number to make the plot smooth.  
( Note: in this course, we suggest set as 1000, which means plotting 1000 points)

```python
# the ?? is the work for you
x = np.linspace(-10, 10, 1000)
y = -x**2 + 9*x + 100
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
```

<div style="border-left: 4px solid #28a745; padding: 10px; background-color: #f9fff3; margin-top: 10px;"><strong>Task 2.2 Plot a time-dependent variable</strong> 
</div>

In this course, we will explore physical systems whose behavior changes over time. 

Mathmatically, we use the **time-dependent variable** to denote their behaviors.

For example:

* Charging a capacitor: voltage, $V$, across a capacitor changes over time. We can denote it as $V(t)$.
* Cooling a room: temperature, $T$, of a room decreases over time. We denote it as $T(t)$.
* Accelerating a car: velocity, $v$, of a car increases over time. We denote it as $v(t)$.
* Radioactive decay: radiation, $N$, from nuclear materials, decays over time. We denote it as $N(t)$.
* ....

And $t$ is set as SI unit (International System of Units) -- second.

Suppose, you are provided the math expression of a time-depedent variable, such as voltage $V(t) = 20e^{-0.5t}$ .  
How to plot it?

The way to plot is very similar as Task 2.1 introduced. We just need to
* Treat the $t$ as X-axis. <br> Note: time typically counts from 0 sec, so the first entry of ``np.linspace(  ,  , )`` typically sets as 0  
* Treat the $V(t)$ as Y-axis

```python
# This is an example plotting the behavior from 0 to 10 seconds.
t = np.linspace(0, 10, 1000)
V_t = 20*np.exp(-0.5*t)
plt.plot(t,V_t)
plt.xlabel("t")
plt.ylabel("V_t")
plt.grid(True)
```
