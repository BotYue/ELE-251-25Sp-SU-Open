A Sine Signal is in such math expression:

$$x(t) = A\sin(\omega t+\phi)$$

or 

$$x(t) = A\sin(2\pi f t+\phi)$$

* $A$: **Amplitude**, or **Magnitude**. The unit depends on your problem setting. $2A$ is called peak-to-peak amplitude
* $\omega$: **radial frequency (radians per second, rad/s)**
* $f$: **Frequency (Hertz, Hz)**
* $T$: **Period (seconds)**. The conversion is $T=\dfrac{1}{f}$
* $\phi$: **Phase Shift (radians, rad)**

**Firstly, many natural phenomena exhibit sinusoidal motion**, such as:

* Sound waves
* Electromagnetic waves
* Electrical AC signals
* Mechanical vibrations
* ...

In the next code piece, it generates a .wave sound file based on a pure Sine Signal.  
You can open and hear this sound file in your computer.

Try to adjust ```frequency = 650``` to other values and experience their differences.

````python
from scipy.io import wavfile
sampling_rate = 44100  # Hz (Default, No Change)

# Define sound parameters
duration = 3  # seconds
frequency = 500  # Hz 

t_sound = np.linspace(0, duration, int(sampling_rate * duration))
waveform = np.sin(2 * np.pi * frequency * t_sound)  

# Save as a WAV file
waveform_int16 = np.int16(waveform * 32767) 
wavfile.write("sine_wave1.wav", sampling_rate, waveform_int16)

print(".wav has been created and saved to your folder")
````

**Secondly, Sine signals are the fundamental building blocks of many periodic signals.**

You can approximately represent other periodic signals as a sum of multiple sine waves. 

**Example 1: Sum sine waves to approximate a square wave** 

First run the code and see the plot,  
Then gradually add more terms to ``signal_sqr`` to observe what is happening.

````python
t_sqr = np.linspace(0, np.pi, 1000)  
omega = 2 * np.pi

signal_sqr = (
    (4 / (np.pi * 1)) * np.sin(1 * omega * t_sqr)
    +(4 / (np.pi * 3)) * np.sin(3 * omega * t_sqr)
    +(4 / (np.pi * 5)) * np.sin(5 * omega * t_sqr)
    +(4 / (np.pi * 7)) * np.sin(7 * omega * t_sqr)
    +(4 / (np.pi * 9)) * np.sin(9 * omega * t_sqr)
    +(4 / (np.pi * 11)) * np.sin(11 * omega * t_sqr)
    +(4 / (np.pi * 13)) * np.sin(13 * omega * t_sqr)
)

plt.plot(t_sqr, signal_sqr)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
````

**Example 2: Sum sine waves to approximate a sawtooth wave** 

First run the code and see the plot,  
Then gradually add more terms to ``signal_saw`` to observe what is happening.

````python
t_saw = np.linspace(0, np.pi, 1000)  
omega = 2 * np.pi

signal_saw = (
    (2 / (np.pi * 1)) * np.sin(1 * omega * t_saw)
    +(2 / (np.pi * 2)) * np.sin(2 * omega * t_saw)
    +(2 / (np.pi * 3)) * np.sin(3 * omega * t_saw)
    +(2 / (np.pi * 4)) * np.sin(4 * omega * t_saw)
    +(2 / (np.pi * 5)) * np.sin(5 * omega * t_saw)
    +(2 / (np.pi * 6)) * np.sin(6 * omega * t_saw)
    +(2 / (np.pi * 7)) * np.sin(7 * omega * t_saw)
)

plt.plot(t_saw, signal_saw)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
````

  <h2 style="margin: 0;">Task 2: Complex Exponential Form of Sine Signals</h2>

The **Euler's formula** states:
$$e^{j\theta} = \cos(\theta) + j\sin(\theta)$$

From this, we can express **sine and cosine** as a **Complex Exponential form**:

$$\sin(\omega t + \phi) = \frac{e^{j(\omega t + \phi)} - e^{-j(\omega t + \phi)}}{2j}$$

$$\cos(\omega t + \phi) = \frac{e^{j(\omega t + \phi)} + e^{-j(\omega t + \phi)}}{2}$$

Let's study this Sine Signal:
$$5\sin(2t+0.3)$$

Create a Markdown below, **type its Complex Exponential form**

<span style="border-left: 4px solid #FFA500; padding: 5px; background-color: #FFF4E5; display: inline-block;">
    <strong>Complex Exponential form: 1 pt</strong>
</span>

Complete the code cell, **plot original sine form and its complex exponential form in the same plot**

Note: 

* A trick to simply typing is using ```np.conj``` . As you can see, $e^{j(\omega t + \phi)}$ and $e^{-j(\omega t + \phi)}$ are in conjugate
* In Python, we type complex numbers as ```1j```, ```2j```, ...

````python
t_task = np.linspace(0, 15, 1000)  

signal_orig_form = 5*np.sin(2*t_task + 0.3)

sine_comp_term1 = np.exp(1j*(2*t_task+0.3))
sine_comp_term2 = np.exp(-1j*(2*t_task+0.3))

signal_exp_form = 5*(sine_comp_term1 - sine_comp_term2)/(2j)

plt.figure(figsize=(12, 5))
plt.plot(t_task, signal_orig_form, '--', label="Original sine form", linewidth=2.5, alpha=0.8)
plt.plot(t_task, signal_exp_form, label="Complex exponential form", linewidth=1)

plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend(loc="upper right")
plt.grid(True)            
````
